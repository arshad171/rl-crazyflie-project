```python
import tensorboard as tb
import tensorflow as tf
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd
import plotly.express as px
```

```python
event_acc_ol = EventAccumulator("results/nav-hover-22/logs", size_guidance={"scalars": 0})
event_acc_ol.Reload()

event_acc_cl = EventAccumulator("results/nav-hover-pid-loop-2/logs", size_guidance={"scalars": 0})
event_acc_cl.Reload()
```

```python
open_loop_r = event_acc_ol.Scalars("train/rewards")
closed_loop_r = event_acc_cl.Scalars("train/rewards")

open_loop_r_df = pd.DataFrame(open_loop_r)
closed_loop_r_df = pd.DataFrame(closed_loop_r)
```

```python
fig_r = px.line()
fig_r.update_layout(
    # title_text="attr",
    # title_xanchor="center",
    # title_x=0.5,
    xaxis_title="Timesteps",
    yaxis_title="Expected Reward",
)

fig_r.add_scatter(x=open_loop_r_df.index * (open_loop_r_df.iloc[1]["step"] - open_loop_r_df.iloc[0]["step"]), y=open_loop_r_df.value, name="RL + Open Loop")
fig_r.add_scatter(x=closed_loop_r_df.index * (closed_loop_r_df.iloc[1]["step"] - closed_loop_r_df.iloc[0]["step"]), y=closed_loop_r_df.value, name="RL + Closed Loop")
```

```python
open_loop_a = event_acc_ol.Scalars("train/actor_loss")
closed_loop_a = event_acc_cl.Scalars("train/actor_loss")

open_loop_a_df = pd.DataFrame(open_loop_a)
closed_loop_a_df = pd.DataFrame(closed_loop_a)
```

```python
fig_a = px.line()
fig_a.update_layout(
    # title_text="attr",
    # title_xanchor="center",
    # title_x=0.5,
    xaxis_title="Timesteps",
    yaxis_title="Expected Reward",
)

fig_a.add_scatter(x=open_loop_a_df.index * (open_loop_a_df.iloc[1]["step"] - open_loop_a_df.iloc[0]["step"]), y=open_loop_a_df.value, name="RL + Open Loop")
fig_a.add_scatter(x=closed_loop_a_df.index * (closed_loop_a_df.iloc[1]["step"] - closed_loop_a_df.iloc[0]["step"]), y=closed_loop_a_df.value, name="RL + Closed Loop")
```
