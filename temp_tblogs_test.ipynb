```python
import tensorboard as tb
import tensorflow as tf
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd
import plotly.express as px
```

```python
event_acc_ppo_kl = EventAccumulator("results/rl-train-test-ppo-kl/logs", size_guidance={"scalars": 0})
event_acc_ppo_kl.Reload()

event_acc_ppo = EventAccumulator("results/rl-train-test-ppo/logs", size_guidance={"scalars": 0})
event_acc_ppo.Reload()

event_acc_td3 = EventAccumulator("results/rl-train-test-td3/logs", size_guidance={"scalars": 0})
event_acc_td3.Reload()
```

```python
ppo_kl = event_acc_ppo_kl.Scalars("train/returns")
ppo = event_acc_ppo.Scalars("train/returns")
td3 = event_acc_td3.Scalars("train/rewards")

ppo_kl_df = pd.DataFrame(ppo_kl)
ppo_df = pd.DataFrame(ppo)
td3_df = pd.DataFrame(td3)
```

```python
fig_r = px.line()
fig_r.update_layout(
    title_text="Training Performance",
    title_xanchor="center",
    title_x=0.5,
    xaxis_title="Timesteps",
    yaxis_title="Expected Reward",
)

fig_r.add_scatter(x=ppo_kl_df.index * (ppo_kl_df.iloc[1]["step"] - ppo_kl_df.iloc[0]["step"]), y=ppo_kl_df.value, name="PPO - ADA KL")
fig_r.add_scatter(x=ppo_df.index * (ppo_df.iloc[1]["step"] - ppo_df.iloc[0]["step"]), y=ppo_df.value, name="PPO - CLIP")
fig_r.add_scatter(x=td3_df.index * (td3_df.iloc[1]["step"] - td3_df.iloc[0]["step"]), y=td3_df.value, name="TD3")
```
